```python
# exercises with sales data
# the original data has error in revenue calculation. 
# Its revenue = profit + cost, but cannot reconcile with unit_price * quantity
# But its cost = unit_cost * quantity

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sales = pd.read_csv('sales_data.csv', parse_dates = ['Date'])
sales.head()
```

```python
sales['Calculated_Cost'] = sales['Order_Quantity'] * sales['Unit_Cost']

sales['Calculated_Cost'].head()
```

```python
# reconciliation

(sales['Calculated_Cost'] != sales['Cost']).sum()
```

```python
# Using scatter to explore relationship between calculated cost and profit

sales.plot(kind='scatter', x = 'Calculated_Cost', y = 'Profit', figsize = (6, 6))
```

```python
# reconcile revenue

sales['Calculated_Revenue'] = sales['Cost'] + sales['Profit']
sales['Calculated_Revenue'].head()
```

```python
(sales['Calculated_Revenue'] - sales['Revenue']).sum()
```
