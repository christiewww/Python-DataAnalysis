```python
# Conditional selection

df['Population'] > 70
```

```python
df[df['Population'] > 70]
```

```python
# in practice we create a 'mask' for conditional selection

mask = df['Population'] > 70
df[mask]
```

```python
# show specific columns with conditional selection

df.loc[df['Population']>70, ['Population', 'GDP', 'HDI']]
```

```python
# show specific columns with more than one condition

df.loc[(df['Population'] > 50) & (df['GDP'] > 3000000), ['Population', 'GDP', 'HDI']]
```

```python
# drop row (there are other ways to do it but this is simpler)

df.drop(['Italy', 'Canada'], axis=0)
```

```python
# drop column (there are other ways to do it but this is simpler)

df.drop(['Surface Area'], axis=1)
```

```python
# Operation in data frome

df[['Population', 'GDP']] / 100
```

```python
# operation with multiple columns

# step 1: to create a series (pandas)
crisis = pd.Series([-1_000_000, -0.3], index = ['GDP', 'HDI'])
crisis
```

```python
# step 2: add the series to the data frame

df[['GDP', 'HDI']] + crisis
```

```python
# add a new column to data frame

# step 1: to create a new series

langs = pd.Series(['French', 'German', 'Italian', 'Janpanese', 'English', 'English'], index = ['France', 'Germany', 'Italy', 'Japan', 'United Kingdom', 'United States'], name ='Language')
langs
```

```python
df['Language'] = langs
df
```
