Data Visualization - Canada Immigrants

```python
import numpy as np
import pandas as pd

df_can = pd.read_excel('Data//Canada.xlsx', sheet_name='Canada by Citizenship', skiprows = range(20), skipfooter = 2)
df_can.head()

# sheet_name
```

```python
df_can.tail()
```

```python
df_can.info()
```

```python
df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
df_can.head(2)
```

```python
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
df_can.head()
```

```python
df_can.set_index('Country', inplace=True)
df_can.head()
```

```python
df_can.columns
```

```python
df_can['Total'] = df_can[[1980,        1981,
              1982,        1983,        1984,        1985,        1986,
              1987,        1988,        1989,        1990,        1991,
              1992,        1993,        1994,        1995,        1996,
              1997,        1998,        1999,        2000,        2001,
              2002,        2003,        2004,        2005,        2006,
              2007,        2008,        2009,        2010,        2011,
              2012,        2013,]].sum(axis=1)
df_can.head()

# syntax in the solution df_can.sum(axis=1) does not work with earning that cannot concat str to int
```

```python
df_can.describe()
```

```python
df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)
top_5 = df_can.head(5)
top_5
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
Haiti = df_can.loc['Haiti']
haiti = pd.DataFrame(Haiti)
haiti = haiti.drop(['Continent', 'Region', 'DevName', 'Total'])
```

```python
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

plt.text(20, 6000, '2010 Earthquake')  # annotation

plt.show()
```

```python
Chi_ind = df_can.loc[['China', 'India']]
chi_ind = pd.DataFrame(Chi_ind)
chi_ind = chi_ind.drop(['Continent', 'Region', 'DevName', 'Total'], axis=1)
chi_ind
```

```python
chi_ind = chi_ind.transpose() #switch columns and rows
chi_ind.head()
```

```python
chi_ind.plot(kind = 'line')

plt.title('Immigrants from China and India')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
df_north = df_can.loc[['Denmark', 'Norway', 'Sweden']]
df_north = df_north.drop(['Continent', 'Region', 'DevName', 'Total'], axis=1)
df_north = df_north.transpose()
df_north.head()
```

```python
df_north.plot(kind = 'bar')

plt.title('Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')

plt.annotate('2008 - 2011 Financial Crisis',  # text to display
             xy=(27, 30),  # start the text at at point (year 2007 , pop 30)
             rotation=45,  # based on trial and error to match the arrow
             va='bottom',  # want the text to be vertically 'bottom' aligned
             ha='left',  # want the text to be horizontally 'left' algned.
             )

plt.show()
```

Pie chart plot

```python
df_continents = df_can.groupby('Continent', axis = 0).sum()
df_continents = df_continents.drop(['Region', 'DevName'], axis = 1)
df_continents
```

```python
fig, ax = plt.subplots(figsize = (4,4))

labels = df_continents.index.values.tolist()
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_list = [0.1, 0, 0, 0, 0.1, 0.1] # ratio for each continent with which to offset each wedge.

ax.pie(df_continents['Total'], 
       labels = labels,
       autopct='%1.1f%%',
       colors = colors_list,
       explode = explode_list)

plt.title('Immigration from Different Continents [1980-2013]')
plt.axis('equal') # Sets the pie chart to look like a circle.
plt.legend(labels=df_continents.index, bbox_to_anchor=(1.3, 1.0), loc='upper left') 
plt.axis('equal')

plt.show()
```

Scatter Plot

```python
df_total = pd.DataFrame(df_can.sum(axis = 0))
df_total = df_total.reset_index()
df_total.columns = ['Year','Total']
df_total = df_total.drop([0,1,2,37], axis = 0)


df_total.head()
```

```python
df_total.plot(kind = 'scatter', x = 'Year', y = 'Total')

plt.title('Total Immigrats to Canada [1980 - 2013]')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')

plt.annotate('dot com bubble burst',  # text to display
             xy=(1999, 150000),  # start the text at at point (year 2008 , pop 30)
             )

plt.show()
```

Subplot Plot

```python
fig = plt.figure(figsize=(10, 10))

ax0 = fig.add_subplot(1, 2, 1) # add subplot 1 (1 row, 2 columns, first plot)
ax1 = fig.add_subplot(1, 2, 2) # add subplot 2 (1 row, 2 columns, second plot)

# Subplot 1: Box plot
chi_ind.plot(kind='box', color='blue', vert=True, figsize=(20, 6), ax=ax0) # add to subplot 1
ax0.set_title('Box Plots of Immigrants from China and India (1980 - 2013)')
ax0.set_xlabel('Countries')
ax0.set_ylabel('Number of Immigrants')

# Subplot 2: Line plot
chi_ind.plot(kind='line', figsize=(20, 6), ax=ax1) # add to subplot 2
ax1.set_title ('Line Plots of Immigrants from China and India (1980 - 2013)')
ax1.set_xlabel('Years')
ax1.set_ylabel('Number of Immigrants')

plt.show()
```

Plot directly with Matplotlib

```python
df_total1 = pd.DataFrame(df_can.sum(axis = 0))
df_total1.columns = ['Total']
df_total1 = df_total1.drop(['Continent', 'Region', 'DevName', 'Total'], axis = 0)

df_total1.head()
```

```python
#Create figure and axes
fig, ax = plt.subplots()

#Changing the index type to integer
df_total1.index = df_total1.index.map(int)

# Customizing the appearance of Plot
ax.plot(df_total1, 
        marker='s', #Including markers in squares shapes
        markersize=5, #Setting the size of the marker
        color='green', #Changing the color of the line
        linestyle="dotted") #Changing the line style to a Dotted line
#Setting up the Title
ax.set_title('Immigrants between 1980 to 2013') 

#Setting up the Labels
ax.set_xlabel('Years')
ax.set_ylabel('Total Immigrants')
ax.legend(['Immigrants'])

#Enabling Grid
plt.grid(True)  #or ax.grid()

#Legend
plt.legend(["Immigrants"]) #or ax.legend()

plt.show()
```

```python
#Create figure and axes
fig, ax = plt.subplots(figsize=(8, 4))

# Customizing Scatter Plot 
ax.scatter(df_total['Year'], df_total['Total'], 
           marker='o', #setting up the markers
           s = 20, #setting up the size of the markers
           color='darkblue')#the color for the marker

plt.title('Immigrants between 1980 to 2013') 
plt.xlabel('Years')
plt.ylabel('Total Immigrants') 

#Legend at upper center of the figure
ax.legend(["Immigrants"], loc='upper center')

#including grid
plt.grid(True)

#Display the plot
plt.show()
```
